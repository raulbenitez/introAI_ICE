Load iris data and split train & test

```python
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y = iris.target
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
# split the data with 70% in each set
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, random_state=0,train_size=0.7)
```

```python
Xtrain.shape
```

Visualitzar dades de training i test

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df_train = pd.DataFrame(Xtrain)
df_train['class'] = ytrain

fig = plt.figure(figsize=(4,4))
sns.pairplot(df_train,hue='class')
```

```python
df_test = pd.DataFrame(Xtest)
df_test['class'] = ytest
fig = plt.figure()
sns.pairplot(df_test,hue='class')
```

Load mpg data and split train & test

```python
import seaborn as sns
dmpg = sns.load_dataset('mpg')
dmpg.head()
```

```python
dmpg.columns
```

```python
dmpg.drop('name',axis=1,inplace=True)
dmpg.head()
```

```python
X = dmpg[['mpg', 'cylinders', 'displacement', 'horsepower', 'weight','acceleration', 'model_year']]
y = dmpg['origin']
```

```python
X.shape
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, random_state=0,train_size=0.7)
```
